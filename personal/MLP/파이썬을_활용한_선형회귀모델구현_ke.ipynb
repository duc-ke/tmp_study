```python
import numpy as np
```

```python
data = np.loadtxt(input, skiprows=1, delimiter=',')
data
```

```python
x = data[:,0]
y = data[:,1]
print('x 데이터 값은: {}\ny 데이터 값은: {}'.format(x, y))
print(f'x 데이터: {x}\ny 데이터 값은: {y}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 7))
plt.plot(x, y, 'o')
```

```python
def scaling(x):
    return (x - x.mean()) / x.std()
```

```python
def hy(x):
    yhat = w0 + w1 * x
    return yhat
```

```python
def mse():
    return (1/len(y)) * sum((y - hy(x))**2)
```

```python
x = scaling(x)
x
```

```python
w0 = np.random.rand()
w1 = np.random.rand()
(w0, w1)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, y, 'o')
plt.plot(x, hy(x))
```

```python
lr = 0.001
diff = 1
error = mse()
while diff > 0.01:
    w1 = w1 + lr*(2/len(y))*sum((y-hy(x))*x)
    w0 = w0 + lr*(2/len(y))*sum(y-hy(x))
    new_error = mse()
    diff = error - new_error
    print(f'weight1:{w1}, weight0:{w0}, cost:{new_error}, diff:{diff}')
    error = new_error
```
