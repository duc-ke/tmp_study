### 데이터 불러 오기

```python
import numpy as np

# study_isg_deeplearning git clone
!git clone https://github.com/duc-ke/study_isg_deeplearning.git
input_file = 'study_isg_deeplearning/heigh_weight.csv'
train = np.loadtxt(input_file, delimiter=',', skiprows=1)
train
```

```python
x_train = train[:, 0]
y_train = train[:, 1]
x_train, y_train, x_train.shape, y_train.shape
```

```python
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
x_train, y_train, x_train.shape, y_train.shape
```

### feature 정규화

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
```

### 모델링 및 학습

```python
from keras.models import Sequential 
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1, activation='linear', input_shape=(1, ))) 
model.compile(loss='mse', optimizer='sgd')
```

```python
model.fit(x_train, y_train, epochs=1000, validation_split=0.2)
```

```python
# 학습된 weight 확인
for layer in model.layers:
    weights = layer.get_weights() # list of numpy arrays
    print(weights)
```

### 학습된 weight로 회귀선 그려서 확인해 보기

```python
y_pred = model.predict(x_train)
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_train, y_train, 'o')
plt.plot(x_train, y_pred)
```
