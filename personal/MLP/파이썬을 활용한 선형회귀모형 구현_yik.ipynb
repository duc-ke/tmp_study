```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def z_score(x):
    return (x-x.mean())/x.std()
```

```python
def hy(x):
    yhat = w0 + w1*x
    return yhat
```

```python
def mse(x,y):
    return (1/n)*np.sum((hy(x) - y)**2) 
```

```python
data = np.loadtxt('data/heigh_weight.csv', skiprows = 1, delimiter = ',')
data
```

```python
x = data[:,0]
y = data[:,1]
```

```python
plt.figure(figsize = (10,7))
plt.plot(x, y, 'o')
```

```python
x = z_score(x)
```

```python
w0 = np.random.rand()
w1 = np.random.rand()
```

```python
plt.figure(figsize =(10,7))
plt.plot(x,y, 'o')
plt.plot(x,hy(x))
```

```python
n = len(y)
n
```

optimizer 구현 

```python
count = 0
lr = 0.001
diff = 1
error = mse(x,y)

while diff > 0: 
    w0 = w0 - lr*(2/n)*np.sum(hy(x) - y)
    w1 = w1 - lr*(2/n)*np.sum((hy(x) - y)*x)
    new_error = mse(x,y)
    diff = error-new_error
    count += 1
    print('{}: w0 = {}, w1 = {}, error = {}, diff = {}'.format(count, w0, w1, error, diff))
    error = new_error 
    
```
